```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
file_one = "Vaccinations_County_Stats_-_Public_Share.csv"
```

```python
file_one_df = pd.read_csv(file_one)
```

```python
file_one_df = file_one_df.dropna(how='any')
```

```python
file_one_df
```

```python
clean_Data = file_one_df.rename(
   columns={"Category":"Category","Category_Description": "Category Description","General_Population":"Population","At_Least_One_Dose":"One Dose","Fully_Vaccinated":"Fully Vaccinated"})
```

```python
clean_Data
```

```python
clean_Data
```

```python
Data_frame = clean_Data.dropna(how='any')
Data_frame
```

```python
data_count =len(Data_frame["Category"].unique())
data_count 
```

```python
Data_frame
```

```python
del Data_frame["Note"]
```

```python
del Data_frame["ObjectId"]
```

```python
X = Data_frame.iloc[:, 3].values.reshape(-1, 1)  # values converts it into a numpy array
Y = Data_frame.iloc[:, 2].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
# Set x axis label
plt.xlabel("Single dose")
# Set y axis label
plt.ylabel("Population")
plt.title( "      The correlation between population and single dose")
plt.show()
```

```python
X = Data_frame.iloc[:, 4].values.reshape(-1, 1)  # values converts it into a numpy array
Y = Data_frame.iloc[:, 2].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
# Set x axis label
plt.xlabel("Fully Vaccinated")
# Set y axis label
plt.ylabel("Population")
plt.title(  "            The correlation between population and Fully Vaccinated")
plt.show()
```

```python
Data_frame["Date"] = [x[:10] for x in Data_frame["Date"].tolist()]
Data_frame
```

```python
One_Dose = Data_frame.groupby('Date').agg({"One Dose":"sum"})
One_Dose
```

```python
category_data = Data_frame.groupby('Category')
category_data.head()
```

```python
dates = Data_frame["Date"].unique()
Fully_Vaccinated = Data_frame.groupby("Date").agg({"Fully Vaccinated":"sum"})
Fully_Vaccinated
```

```python
plt.plot(dates, One_Dose["One Dose"],color="red", label="One Dose")
plt.plot(dates, Fully_Vaccinated["Fully Vaccinated"],color="blue", label="Fully Vaccinated")
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.title("one dose/Fully Vaccinated over time ")
plt.xlabel("Date")
plt.ylabel("people vaccinted")
plt.show()
```

```python
age_groups_df = Data_frame.loc[Data_frame["Category"] == "5Cat Age Groups"]
age_groups = age_groups_df["Category Description"].unique()
group_sum = [age_groups_df[age_groups_df["Category Description"] == x]["Fully Vaccinated"].sum() for x in age_groups]
group_sum
```

```python
group_sum
x_axis = np.arange(len(group_sum))
```

```python
tick_locations = [value for value in x_axis]
plt.bar(age_groups, group_sum, color='r', alpha=0.5, align="center")
plt.title("Age Group/Fully vaccinated ")
plt.xlabel("Age Group")
plt.ylabel("Number of Fully vaccinated people")
plt.show()
```

```python
one_dose_sum = [age_groups_df[age_groups_df["Category Description"] == x]["One Dose"].sum() for x in age_groups]
one_dose_sum
```

```python
tick_locations = [value for value in x_axis]
plt.bar(age_groups, one_dose_sum, color='b', alpha=0.5, align="center")
plt.title("Age Group/one dose vaccinated ")
plt.xlabel("Age Group")
plt.ylabel("Number of one dose vaccinated people")
```
